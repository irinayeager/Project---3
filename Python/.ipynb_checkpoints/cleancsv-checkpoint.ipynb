```python
import pandas as pd
from colors import rgb, hex
import numpy as np
```

```python
df = pd.read_json("../Data/test.json", "r", encoding="utf8")
df.head()
```

```python
df['hex'] = df['hex'].apply(lambda x: tuple(hex(x).rgb))
```

```python
df['hex'][0][0]
```

```python
df['hex'][0][1]
```

```python
df['hex'][0][2]
```

```python
red = []
green = []
blue = []

for i in df['hex']:
    red.append(i[0])
    green.append(i[1])
    blue.append(i[2])
```

```python
df['red'] = red
df['green'] = green
df['blue'] = blue
```

```python
df
```

```python
example = (222.24678456591653 + 193.00401929260468 + 181.0803858520894)
```

```python
df['test'] = example
df
```

```python
df['rgbsum'] = df['red'] + df['green'] + df['blue']
df
```

```python
df['difference'] = np.abs(df['rgbsum'] - df['test'])
df
```
